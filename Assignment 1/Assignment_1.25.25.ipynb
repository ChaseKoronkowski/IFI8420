#### E. Plot the monthly prices of three of the stocks. Display the price series together in one plot (showing the trend together) and also separately. Label the plots correctly   (6 points)

````python
stocks = {"Apple": "AAPL", "Gold": "NEM", "Nvidia": "NVDA"}
 
monthly_data = {
    stock: data[stock][('Close', ticker)].resample('ME').last() for stock, ticker in stocks.items()
}
 
plt.figure(figsize=(12, 6))
for stock in stocks:
    monthly_data[stock].plot(label=stock)
 
plt.title("Monthly Closing Prices: Apple, Gold, and Nvidia")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
 
for stock in stocks:
    plt.figure(figsize=(12, 6))
    monthly_data[stock].plot(label=f"{stock}'s Monthly Closing Price", color='blue')
    plt.title(f"{stock}'s Monthly Closing Price")
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.show()
````

#### F. Compute log returns using adjusted closing price. Write the program efficiently so you can compute all the three stocks returns using functions, loops etc.   (8 points)

````python
# Note: Chase I know you was working on this, update accordingly. 
def compute_log_returns():
    aapl_daily_log_returns = np.log(apple_df['Close']/ apple_df['Close'].shift(1))
    nem_daily_log_returns = np.log(gold_df['Close']/ gold_df['Close'].shift(1))
    nvda_daily_log_returns = np.log(nvidia_df['Close']/ nvidia_df['Close'].shift(1))
    
    return aapl_daily_log_returns, nem_daily_log_returns, nvda_daily_log_returns
````

#### G. Construct time plot of monthly log returns of the three stock for the sample data. Display the return series together in one plot and also separately. Label the plots correctly   (6 points)

````python
# Compute log returns
aapl_log_return, nem_log_return, nvda_log_return = compute_log_returns()

# Convert to DataFrame and resample to monthly frequency
aapl_monthly = aapl_log_return.resample('ME').mean()
nem_monthly = nem_log_return.resample('ME').mean()
nvda_monthly = nvda_log_return.resample('ME').mean()

# Plot combined return series
plt.figure(figsize=(12, 6))
plt.plot(aapl_monthly, label='Apple', color='blue')
plt.plot(nem_monthly, label='Gold', color='orange')
plt.plot(nvda_monthly, label='Nvidia', color='green')
plt.xlabel('Date')
plt.ylabel('Log Returns')
plt.title('Monthly Log Returns of Apple, Gold, and Nvidia')
plt.legend()
plt.grid(True)
plt.show()

# Plot each return series separately
fig, axes = plt.subplots(3, 1, figsize=(12, 18))

# Apple
axes[0].plot(aapl_monthly, label='Apple', color='blue')
axes[0].set_xlabel('Date')
axes[0].set_ylabel('Log Returns')
axes[0].set_title('Monthly Log Returns of Apple')
axes[0].legend()
axes[0].grid(True)

# Gold
axes[1].plot(nem_monthly, label='Gold', color='orange')
axes[1].set_xlabel('Date')
axes[1].set_ylabel('Log Returns')
axes[1].set_title('Monthly Log Returns of Gold')
axes[1].legend()
axes[1].grid(True)

# Nvidia
axes[2].plot(nvda_monthly, label='Nvidia', color='green')
axes[2].set_xlabel('Date')
axes[2].set_ylabel('Log Returns')
axes[2].set_title('Monthly Log Returns of Nvidia')
axes[2].legend()
axes[2].grid(True)

plt.tight_layout()
plt.show()
````

#### H. Compute the sample mean, standard deviation, minimum, and maximum of the log return series (4 points).

````python
# Function to compute statistics
def compute_statistics(df):
    mean = df.mean()
    std_dev = df.std()
    min_value = df.min()
    max_value = df.max()
    return mean, std_dev, min_value, max_value

# Compute statistics for each stock
apple_stats = compute_statistics(apple_df)
gold_stats = compute_statistics(gold_df)
nvidia_stats = compute_statistics(nvidia_df)

# Display the results
print("Apple Statistics:")
print(f"Mean: {apple_stats[0]}")
print(f"Standard Deviation: {apple_stats[1]}")
print(f"Minimum: {apple_stats[2]}")
print(f"Maximum: {apple_stats[3]}\n")

print("Gold Statistics:")
print(f"Mean: {gold_stats[0]}")
print(f"Standard Deviation: {gold_stats[1]}")
print(f"Minimum: {gold_stats[2]}")
print(f"Maximum: {gold_stats[3]}\n")

print("Nvidia Statistics:")
print(f"Mean: {nvidia_stats[0]}")
print(f"Standard Deviation: {nvidia_stats[1]}")
print(f"Minimum: {nvidia_stats[2]}")
print(f"Maximum: {nvidia_stats[3]}\n")
````
