#### E. Plot the monthly prices of three of the stocks. Display the price series together in one plot (showing the trend together) and also separately. Label the plots correctly   (6 points)

````python
stocks = {"Apple": "AAPL", "Gold": "NEM", "Nvidia": "NVDA"}
 
monthly_data = {
    stock: data[stock][('Close', ticker)].resample('ME').last() for stock, ticker in stocks.items()
}
 
plt.figure(figsize=(12, 6))
for stock in stocks:
    monthly_data[stock].plot(label=stock)
 
plt.title("Monthly Closing Prices: Apple, Gold, and Nvidia")
plt.xlabel("Date")
plt.ylabel("Closing Price (USD)")
plt.legend()
plt.grid(True)
plt.show()
 
for stock in stocks:
    plt.figure(figsize=(12, 6))
    monthly_data[stock].plot(label=f"{stock}'s Monthly Closing Price", color='blue')
    plt.title(f"{stock}'s Monthly Closing Price")
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.show()
````
